```python
import community
import networkx as nx
import matplotlib.pyplot as plt

#better with karate_graph() as defined in networkx example.
#erdos renyi don't have true community structure
#G = nx.erdos_renyi_graph(30, 0.05)

G=nx.karate_club_graph()

#first compute the best partition
partition = community.best_partition(G)

#drawing
size = float(len(set(partition.values())))
nodec = ['r','b','g','k']
pos = nx.spring_layout(G)
count = 0.
for com in set(partition.values()) :
    count = count + 1.
    list_nodes = [nodes for nodes in partition.keys()
                                if partition[nodes] == com]
    nx.draw_networkx_nodes(G, pos, list_nodes, node_size = 40,node_color =nodec[com])
                                #node_color = str(count / size))


nx.draw_networkx_edges(G,pos, alpha=0.5)
plt.show()
```

```python
L = ['b', 'c', 'e', 'f']
g = nx.complete_graph(len(L))
nx.relabel_nodes(g,dict(enumerate(L)), copy = False)
g.add_edges_from([('a','b'),('a','f'),('d','c'),('d','e')])
cl = list(nx.find_cliques(g))
nx.draw(g)
plt.show()
print cl
```

```python
max_cliques = list(nx.find_cliques(g)) #outputs list of lists of cliques
for clique in max_cliques:
    mygraph = nx.complete_graph(len(clique))
    nx.relabel_nodes(mygraph,dict(enumerate(clique)), copy = False)
    nx.draw(mygraph)
    plt.show()
```

```python
from networkx.algorithms import community
G = nx.barbell_graph(5, 1)
communities_generator = community.girvan_newman(G)
top_level_communities = next(communities_generator)
next_level_communities = next(communities_generator)
sorted(map(sorted, next_level_communities))
```

```python
import itertools
G1=nx.karate_club_graph()
com = community.girvan_newman(G1)
k=2
for communities in itertools.islice(com, k):
    print(tuple(sorted(c) for c in communities))
#tuple(sorted(c) for c in next(K))
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from gridworld import GridWorldExample3
from rlagents import RLExampleAgent, DP_Agent, TDSarsa_Agent
from plotFunctions import plotWorld, plotStateValue, plotStateActionValue, plotPolicyPi, plotGreedyPolicyQ
```

```python
# build and plot world
world = GridWorldExample3()
plotWorld(world,plotNow=False)
plt.show()
```

# Deterministic Policy 

```python
# initialise agent
agent = RLExampleAgent(world)
```

```python
plotStateValue(agent.v, world, plotNow=False)
plt.show()
```

```python
plotStateActionValue(agent.q, world, plotNow=False)
plt.show()
```

```python
plotPolicyPi(agent.P_pi, world, plotNow=False)
plt.show()
```

# Dynamic Programming

<img src="images/greedy.png",width=600,height=800>

<img src="images/value_update.png",width=600,height=800>

```python
agent = DP_Agent(world)
agent.initRandomPolicy()
plotPolicyPi(agent.P_pi, world, plotNow=False)
plotStateValue(agent.v, world, plotNow=False)
plt.show()
```

```python
agent.evaluatePolicy(gamma = 0.9)
plotStateValue(agent.v, world, plotNow=False)
plt.show()
```

```python
agent.improvePolicy()
plotPolicyPi(agent.P_pi, world, plotNow=False)
plt.show()
```

```python
agent.policyIteration(gamma = 0.9)
plotStateValue(agent.v, world, plotNow=False)
plt.show()
```
